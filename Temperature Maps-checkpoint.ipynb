```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_temperature_vs_time(sample_name, initial_temp, final_temp, ramp_rate, time_duration, cool_down_time, color='blue', save_individual=False, individual_path=None):
    fig, ax = plt.subplots()
    
    if ramp_rate is not None:
        # Ramp-up phase
        ramp_time = (final_temp - initial_temp) / ramp_rate
        constant_temp_time = time_duration - ramp_time
        ramp_time_points = np.linspace(0, ramp_time, 100)
        constant_temp_time_points = np.linspace(ramp_time, time_duration, 100)

        ramp_temp_points = np.linspace(initial_temp, final_temp, 100)
        constant_temp_points = np.full(100, final_temp)
    else:
        # No ramp-up phase
        ramp_time = 0
        constant_temp_time = time_duration
        constant_temp_time_points = np.linspace(0, time_duration, 100)

        constant_temp_points = np.full(100, final_temp)

    # Cool-down phase
    cool_down_initial_time = time_duration
    cool_down_final_time = time_duration + cool_down_time
    cool_down_time_points = np.linspace(cool_down_initial_time, cool_down_final_time, 100)

    # Exponential decay for cooling
    tau = cool_down_time / 5  # Decay constant
    cool_down_temp_points = final_temp * np.exp(-(cool_down_time_points - cool_down_initial_time) / tau) - 121

    if ramp_rate is not None:
        # Concatenate all phases with ramp-up
        time_points = np.concatenate((ramp_time_points, constant_temp_time_points, cool_down_time_points))
        temperature_points = np.concatenate((ramp_temp_points, constant_temp_points, cool_down_temp_points))
    else:
        # Concatenate all phases without ramp-up
        time_points = np.concatenate((constant_temp_time_points, cool_down_time_points))
        temperature_points = np.concatenate((constant_temp_points, cool_down_temp_points))

    label = f"{sample_name}: Final Temp {final_temp}°C"
    ax.plot(time_points, temperature_points, label=label, color=color)
    ax.set_xlabel('Time (min)')
    ax.set_ylabel('Temperature (°C)')
    ax.legend()

    if save_individual:
        if individual_path is not None:
            plt.savefig(individual_path, dpi=1200, bbox_inches='tight')
        else:
            plt.savefig(f"{sample_name}.png", dpi=1200, bbox_inches='tight')
    
    return ax.get_lines()
```

```python
# Sample data
line1 = plot_temperature_vs_time('Sample 1', 900, 900, ramp_rate=None, time_duration=30, cool_down_time=100, color='blue', save_individual=True, individual_path='sample1.png')
line2 = plot_temperature_vs_time('Sample 2', 21, 900, ramp_rate=9, time_duration=180, cool_down_time=100, color='red', save_individual=True, individual_path='sample2.png')
line3 = plot_temperature_vs_time('Sample 3', 21, 700, ramp_rate=7, time_duration=300, cool_down_time=100, color='green', save_individual=True, individual_path='sample3.png')

# Save combined plot
fig, ax = plt.subplots()
for line in line1 + line2 + line3:
    ax.plot(line.get_xdata(), line.get_ydata(), label=line.get_label(), color=line.get_color())
ax.set_xlabel('Time (min)')
ax.set_ylabel('Temperature (°C)')
ax.legend()
plt.savefig('combined_samples.png', dpi=1200, bbox_inches='tight')
plt.show()
```
